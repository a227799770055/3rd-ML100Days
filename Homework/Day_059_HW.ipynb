### 用 digits (3個數字即可 , dataset.load_digits(n_class=3)) 資料嘗試跑 PCA


```python
from sklearn import datasets
from sklearn import decomposition
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np 


digits = datasets.load_digits(n_class=3)
X = digits.data
y = digits.target
```

載入 相關套件 並 執行 PCA 實驗 ...

```python
PCA=decomposition.PCA(n_components=63)
X_PCA=PCA.fit_transform(X)
X_PCA.shape
```

```python
fig=plt.figure(figsize=(10,10))
ax=plt.axes()
plt.xlabel("PCA_1",size=20)
plt.ylabel("PCA_2",size=20)
ax.scatter(X_PCA[:,0],X_PCA[:,1], c=y)
ax.text(X_PCA[y==0,0].mean(),X_PCA[y==0,1].mean(),'0',alpha=100,color='red',size=50)
ax.text(X_PCA[y==1,0].mean(),X_PCA[y==1,1].mean(),'1',alpha=100,color='red',size=50)
ax.text(X_PCA[y==2,0].mean(),X_PCA[y==2,1].mean(),'2',alpha=100,color='red',size=50)
```

```python
fig = plt.figure(1, figsize=(10,10))
plt.clf()
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=30, azim=130)
ax.scatter(X_PCA[:,0],X_PCA[:,1],X_PCA[:,2],c=y)
ax.set_xlabel('PCA_1')
ax.set_ylabel('PCA_2')
ax.set_zlabel('PCA_3')
for i in range(3):
    ax.text3D(X_PCA[y==i,0].mean(),X_PCA[y==i,1].mean(),X_PCA[y==i,2].mean(),i,size=30,color='red')
```

```python
PCA.explained_variance_ratio_
plt.figure(figsize=(15,10))
plt.plot(range(63),PCA.explained_variance_ratio_)
plt.xlabel('n_component',size=20)
plt.ylabel('variance_ratio',size=20)
plt.xticks(size=10)
plt.yticks(size=10)
```
