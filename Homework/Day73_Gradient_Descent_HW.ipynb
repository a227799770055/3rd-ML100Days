# 作業重點

使用function y=(x+5)² 來練習學習率( leanrning rate ) 調整所造成的影響

# 作業目標:
    請嘗試使用不同的組合驗證 learning rate 對所需 iteration 的影響
    lr = [0.1, 0.0001]
    主要驗證 Lr 對於grandient 收斂的速度
    

```python
cur_x = 3 # The algorithm starts at x=3
precision = 0.000001 #This tells us when to stop the algorithm
previos_step_size=1
max_iters = 10000 # maximum number of iterations
iters=0
df = lambda x: 2*(x+5) #Gradient of our function 
#Learning Rate is 0.0001
LR=0.0001
histortical_x=[cur_x]
historical_iters=[iters]
while previos_step_size> precision and iters< max_iters:
    pre_x=cur_x
    cur_x=cur_x-LR*(df(cur_x))
    previos_step_size=abs(cur_x-pre_x)
    iters+=1
    histortical_x.append(cur_x)
    historical_iters.append(iters)
    print('current_x={}'.format(pre_x))
    print('iters_number={}'.format(iters))
    
```

```python
print('x={}, precision={}'.format(cur_x, previos_step_size))
```

```python
import matplotlib.pyplot as plt 
plt.plot(historical_iters, histortical_x,'b')
plt.xlabel('iters')
plt.ylabel('X')
plt.title('Learn Rate= 0.0001')
plt.show()

```
